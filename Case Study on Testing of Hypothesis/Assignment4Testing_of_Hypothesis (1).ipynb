**1. The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.**

**one-sample t-test on  difference in sales for each observation** #one method

Null Hypothesis (H0): There is no significant difference between the average increase in sales after the introduction of digital marketing and zero.

Alternative Hypothesis (H1): There is a significant increase in sales after the introduction of digital marketing.

```python
from scipy.stats import ttest_1samp

sales_before_digital = data['Sales_before_digital_add(in $)']                           # Extract sales data before and after digital marketing
sales_after_digital = data['Sales_After_digital_add(in $)']

sales_differences = [after - before for before, after in zip(sales_before_digital, sales_after_digital)] # Calculate differences in sales

t_stat, p_value = ttest_1samp(sales_differences, 0)                                     # Perform one-sample t-test

print("One-Sample t-test results:")
print("t-statistic:", t_stat)                                                           # Output results
print("p-value:", p_value)
                                                                                        # Check significance
if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant increase in average sales after the introduction of digital marketing.")
else:
    print("Fail to reject the null hypothesis. There is no significant increase in average sales after the introduction of digital marketing.")
```

**Two-sample T-test**    #second method

Null Hypothesis (H0): There is no difference in average sales before and after the introduction of digital marketing.

Alternative Hypothesis (H1): There is a significant difference in average sales before and after the introduction of digital marketing.

```python
from scipy.stats import ttest_rel

sales_before_digital = data['Sales_before_digital_add(in $)']        # Extract sales data before and after digital marketing
sales_after_digital = data['Sales_After_digital_add(in $)']

t_stat, p_value = ttest_rel(sales_after_digital,sales_before_digital)# Perform paired t-test

print("Paired t-test results:")                                      # Print the results
print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value < 0.05:                                                   # Check significance
    print("Reject the null hypothesis. There is a significant difference in average sales before and after digital marketing.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in average sales before and after digital marketing.")
```

**2. The company needs to check whether there is any dependency between
thefeatures Region and Manager.**

Null Hypothesis (H0): There is no association between the "Region" and "Manager" features.

Alternative Hypothesis (H1): There is an association between the "Region" and "Manager" features.

**chi-square test**

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(data['Region'], data['Manager'])       # Create a contingency table

chi2, p, dof, expected = chi2_contingency(contingency_table)           # Perform chi-square test

print("Chi-square Statistic:", chi2)                                   # Print results
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)
                                                                       # Check significance
if p < 0.05:
    print("Reject the null hypothesis. There is a significant association between Region and Manager.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between Region and Manager.")
```
