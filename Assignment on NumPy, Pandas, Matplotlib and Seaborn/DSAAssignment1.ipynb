**5. Perform meaningful visualizations using the dataset. Bring at least 3
visualizations.**

```python
import numpy as np
import pandas as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**1.HISTOGRAM**

```python
iris_data.hist(edgecolor='black',linewidth=1)
plt.suptitle('Histograms on iris features.')
plt.show()
```

**2.SCATTER PLOT**

```python
sns.scatterplot(data=iris_data, x='SL', y='SW', hue='Classification')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
sns.scatterplot(data=iris_data, x='PL', y='PW', hue='Classification')
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.show()
```

**3.BOXPLOT**

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.boxplot(x='Classification', y = 'SL', data=iris_data,hue = 'Classification')
plt.subplot(2,2,2)
sns.boxplot(x='Classification', y = 'SW', data=iris_data,hue = 'Classification')
plt.subplot(2,2,3)
sns.boxplot(x='Classification', y = 'PL', data=iris_data,hue = 'Classification')
plt.subplot(2,2,4)
sns.boxplot(x='Classification', y = 'PW', data=iris_data,hue = 'Classification')
plt.suptitle("Box Plot of Iris Dataset")
```

**4.VIOLIN PLOT**

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Classification', y = 'SL', data=iris_data,hue = 'Classification')
plt.subplot(2,2,2)
sns.violinplot(x='Classification', y = 'SW', data=iris_data,hue = 'Classification')
plt.subplot(2,2,3)
sns.violinplot(x='Classification', y = 'PL', data=iris_data,hue = 'Classification')
plt.subplot(2,2,4)
sns.violinplot(x='Classification', y = 'PW', data=iris_data,hue = 'Classification')
plt.suptitle("Violin Plot of Iris Dataset")
```

**5.PAIRPLOT**

```python
plt.figure(figsize=(20,15))
sns.pairplot(iris_data,hue = 'Classification',height=4)
plt.suptitle("Pair Plot of Iris Dataset")
plt.show()
```

**6.HEATMAP**

```python
slice_data=iris_data[['SL','SW','PL','PW']]
```
